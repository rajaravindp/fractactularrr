# **MandelBrot Set**



### **Visual 1:**
**Generate a coarse representation of the Mandelbrot Set using a limited number of 'x' values. The purpose here is to quickly show whether the Mandelbrot sequence remains within a certain range for a specific set of 'n' values. It provides a broad overview of whether the sequence is finite or not for different 'n' values.**

```python
# Initial values for 'n' and 'z'
n = -3
z = 0
```

For n=0 -> 0, 0, 0, 0, 0... constant <br>
For n=1 -> 1, 2, 5, 26, 677... burgeons <br>
For n=-1 -> -1, 0, -1, 0, -1... finite <br>
For n=-2 -> -2, 2, 2, 2, 2... finite <br>
For n=-3 -> -3, 6, 33, 1086, 1179393... burgeons <br>

```python
# Calculate the first iteration of the Mandelbrot sequence for 'n' and store it in 'z'
z = z*z + n
z
```

```python
# Import required libraries
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create array of 'x' values ranging from -3 to 2
x = np.arange(-3, 3, 1)
# Initialize array 'z' to store Mandelbrot sequence vals for 'n' over 10 iterations
z = np.zeros_like(x)
# Create empty array [arr] to store the Mandelbrot seq vals for each iteration
arr = [x]
# Iterate 10 times to calculate and store the Mandelbrot seq vals
for i in range(10):
  z = z*z + n
  arr.append(np.abs(z)) # absolute value of the result is stored in [arr]
```

```python
arr
```

```python
# Print Mandelbrot seq vals for 'n' over the iters
for j in arr:
  print("\t\t".join(map(str, j)))
```

```python
# Creating more granular array of 'x' values -> range= (-3 to 2) & step= 0.01
x = np.arange(-3, 3, 0.01)
# Initialize array 'z' to store the Mandelbrot seq vals for 'x' over 100 iters
z = np.zeros_like(x)
# Create empty array arr to store the Mandelbrot seq vals for each iter
arr = [x]
# Iterate 100 times to calc & store the Mandelbrot seq vals
for i in range(100):
  z = z*z + x   # MandelBrot Equation
  arr.append(np.abs(z) < 10)
# This loop iterates 100 times and calculates the next iteration of the Mandelbrot sequence at each step.
# It also checks if the absolute value of the result is less than 10.
# If it is, the value True is stored in arr, otherwise the value False is stored in arr.
# This is because values of z that are greater than 10 are considered to have escaped the Mandelbrot Set.
```

```python
# Create viz to check if the Mandelbrot seq stays within certain range
figure(figsize=(10, 10))
# Display the array [arr] as an img
imshow(arr, extent=(x.min(), x.max(), 0, 10))
# Display colorbar on the side
colorbar()
```

```python
# Boundaries of the region where the sequence remains within the range
# Min
print(min(x[arr[-1]]))
# Max
print(max(x[arr[-1]]))
# Range - (-1.9900000000000215 - 0.24999999999993072) -> Finite
# For all other numbers -> num not part of the MandelBrot Set.
```

## **Visual 2:**
**Detailed representation of the Mandelbrot Set using a finer granularity of 'x' values. Allows for a closer examination of the boundary between the set and non-set regions. The primary goal here is to find the exact range of 'x' values where the Mandelbrot sequence remains finite and to calculate the minimum and maximum values within that range.**

```python
# Define range for 'x' and 'y' axes in complex plane
x, X = -2, 1
y, Y = -1, 1

# Granularity for grid of complex nums
delta = 0.001
real_line, imaginary_line = np.mgrid[x:X:delta, y:Y:delta] # real_line and imaginary_line - contain grid of real and imaginary numbers respectively
c = (real_line + 1j*imaginary_line).reshape(imaginary_line.shape[0], -1).T # create complex array. Reshape c - same dims as imaginary_line
z = np.zeros_like(c) # Initialize with zeroes. Store Mandelbrot Set vals for each complex number in array [c]

# Iterate 100 times to calculate and visualize the Mandelbrot Set
for i in range(100):
  z = z*z + c # MandelBrot Equation

# Viz Mandelbrot Set
imshow(np.abs(z), extent=(x, X, y, Y))
```

```python
# Define the range for the 'x' and 'y' axes in the complex plane
x, X = -2, 1
y, Y = -1, 1

delta = 0.001
real_line, imaginary_line = np.mgrid[x:X:delta, y:Y:delta]
c = (real_line + 1j*imaginary_line).reshape(imaginary_line.shape[0], -1).T
z = np.zeros_like(c)
esc = np.zeros_like(np.abs(c))

for i in range(100):
  # Perform 1 iter of the Mandelbrot Equation
  z = z*z + c # MandelBrot Equation
  # Check if abs val of 'z' exceeds 4 & it hasn't escaped before
  idx = (np.abs(z) > 4) & (esc == 0)
  # Record the current iter cnt 'i' for newly escaped pts
  esc[idx] = i

# Viz Mandelbrot Set
imshow(esc, extent=(x, X, y, Y))
```

 ## Visual 3:
 Define a function - plot_mandelbrot - to generate & display the Mandelbrot Set within a specified range in the complex plane. Designed to provide flexible way to visualize the set at different zoom levels or around specific user-defined points. This function allows for interactive exploration of the Mandelbrot Set with varying levels of detail.

 ## Happy Exploring!

```python
# Def func to viz MandelBrot in user-defined range
def plot_mandelbrot(x, X, y, Y):
  """

  It calculates escape times for points in the complex plane and displays the set using color mapping.

  """

  # Calculate step size for creating grid of i
  delta = (X-x)/300
  # Create grid of real & imaginary parts of i
  real_line, imaginary_line = np.mgrid[x:X:delta, y:Y:delta]
  # Convert grid into 2D arr of i
  c = (real_line + 1j*imaginary_line).reshape(imaginary_line.shape[0], -1).T
  # Initialize arrays 'z' & 'esc' for Mandelbrot Set calcs & for recording escape time
  z = np.zeros_like(c)
  esc = np.zeros_like(np.abs(c))

  # Iterate 100 times to calculate escape time for each point
  for i in range(100):
    # MandelBrot Equation
    z = z*z + c
    # Check if abs val of 'z' exceeds 4 & it hasn't escaped before
    idx = (np.abs(z) > 4) & (esc == 0)
    # Record current iteration cnt 'i' for newly escaped points
    esc[idx] = i

  # Display Mandelbrot set
  imshow(esc, extent=(x, X, y, Y))
```

```python
# Call the func
plot_mandelbrot(-2, 1, -1, 1)
```

```python
# Convenience function to visualize at user-defined range (by specifying center point (x, y) and delta value 'd')
def plot_mandelbrot_user_points(x, y, d):
  return plot_mandelbrot(x-d, x+d, y-d, y+d)
```

```python
# Call fuunc
plot_mandelbrot_user_points(-0.5, 0.55, 0.1)
```
